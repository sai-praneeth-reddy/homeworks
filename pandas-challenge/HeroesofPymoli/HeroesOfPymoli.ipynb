```python
import pandas as pd
```

```python
# Location of the csv file

file = "Resources/purchase_data.csv"

# Read in the csv file and create a Data Frame

pymoli_df = pd.read_csv(file)

# Get same records from the Data Frame

pymoli_df.head()
```

```python
# Get the unique list of Players

players = pymoli_df['SN'].unique()

# Getting the count of the palyers

total_players = len(players)

# Output the Result to Data Frame

output1 = pd.DataFrame({"Total Players": [total_players]})

output1
```

```python
# Number of Unique Items

items = pymoli_df['Item Name'].unique()

num_uniq_items = len(items)

# Average Purchase Price

pymoli_df.loc[:,"Price"] = pymoli_df["Price"].astype(float)

average_price = pymoli_df['Price'].mean()

average_price_format = "${:.2f}".format(average_price)

# Total Number of Purchases

num_of_purchases = pymoli_df['Purchase ID'].count()

# Total Revenue

tot_rev = pymoli_df['Price'].sum()

tot_rev_format = "${:,}".format(tot_rev)

tot_rev_format

# Output the results to Data Frame

output2 = pd.DataFrame([{"Number of Unique Items": num_uniq_items, "Average price": average_price_format, "Number of Purchases": num_of_purchases, "Total Revenue": tot_rev_format}])

output2
```

```python
# Remove Duplicates

pymoli_unique_df = pymoli_df.drop_duplicates(subset=['SN'])

# Group By Gender

pymoli_group_gender = pymoli_unique_df.groupby(['Gender'])

pymoli_df_gender_count = pymoli_group_gender.count()

# Get the Count

pymoli_df_gender_2 = pymoli_df_gender_count[["Purchase ID"]]

pymoli_df_gender_2 = pymoli_df_gender_2.rename(columns={"Purchase ID": "Total Count"})

# Get the Percentages

pymoli_df_gender_2["Percentage of Players"] = (pymoli_df_gender_2["Total Count"]/total_players)*100

pymoli_df_gender_2["Percentage of Players"] = pymoli_df_gender_2["Percentage of Players"].map("{:.2f}%".format)

output3 = pymoli_df_gender_2

# Output the Results

output3
```

```python
# Group by Gender

pymoli_group = pymoli_df.groupby(['Gender'])

# Get the Count

pymoli_count = pymoli_group.count()

pymoli_count["Purchase Count"] = pymoli_group["Purchase ID"].count()

pymoli_count = pymoli_count[["Purchase Count"]]


# Get Average Purchase Price

pymoli_avg = pymoli_group.mean()

pymoli_avg["Average Purchase Price"] = pymoli_group["Price"].mean()

pymoli_avg = pymoli_avg[["Average Purchase Price"]]


# Get Total Purchase Value

pymoli_total = pymoli_group.sum()

pymoli_total["Total Purchase Value"] = pymoli_group['Price'].sum()

pymoli_total = pymoli_total[["Total Purchase Value"]]


# Get Average Purchase Total Per Person by Gender

pymoli_total["Avg Total Purchase per Person"] = pymoli_total["Total Purchase Value"]/output3["Total Count"]

# Merge all the 3 Tables

pymoli_combine = pd.merge(pd.merge(pymoli_count, pymoli_avg, on="Gender"),pymoli_total, on="Gender")


# Format the Table

pymoli_combine["Average Purchase Price"] = pymoli_combine["Average Purchase Price"].map("${:.2f}".format)

pymoli_combine["Total Purchase Value"] = pymoli_combine["Total Purchase Value"].map("${:.2f}".format)

pymoli_combine["Avg Total Purchase per Person"] = pymoli_combine["Avg Total Purchase per Person"].map("${:.2f}".format)

output4 = pymoli_combine

# Output the Results

output4
```

```python
# Bin the Age Variable

bins = [0, 10, 15, 20, 25, 30, 35, 40, 46]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

pymoli_df["Age Group"] = pd.cut(pymoli_df["Age"], bins, labels=group_names, right=False)

pymoli_unique_df["Age Group"] = pd.cut(pymoli_unique_df["Age"], bins, labels=group_names, right=False)


# Group by Age Group

pymoli_df_age_group_temp = pymoli_unique_df.groupby(["Age Group"])

# Get the count

pymoli_df_age_group_temp = pymoli_df_age_group_temp.count()

pymoli_df_age_group_temp = pymoli_df_age_group_temp[["Purchase ID"]]

pymoli_df_age_group_temp = pymoli_df_age_group_temp.rename(columns={"Purchase ID": "Total Count"})


# Get the Percentage of Players

pymoli_df_age_group_temp["Percentage of Players"] = (pymoli_df_age_group_temp["Total Count"]/total_players)*100


#####################################################################################################################

# Group by Age Group

pymoli_df_age_group = pymoli_df.groupby(["Age Group"])

#Get the Purchase Count, Average Purchase Price, Total Purchase Value

#Count
pymoli_age_demo1 = pymoli_df_age_group.count()

pymoli_age_demo1["Purchase Count"] = pymoli_df_age_group["Purchase ID"].count()

#Average
pymoli_age_demo2 = pymoli_df_age_group.mean()

pymoli_age_demo2["Average Purchase Price"] = pymoli_df_age_group["Price"].mean()


# Total
pymoli_age_demo3 = pymoli_df_age_group.sum()

pymoli_age_demo3["Total Purchase Value"] = pymoli_df_age_group["Price"].sum()

pymoli_age_demo3["Avg Total Purchase per Person"] = pymoli_age_demo3["Total Purchase Value"]/pymoli_df_age_group_temp["Total Count"]


# Merge and Retain

pymoli_age_final = pd.merge(pd.merge(pymoli_age_demo1, pymoli_age_demo2, on="Age Group"),pymoli_age_demo3, on="Age Group")

pymoli_age_final = pymoli_age_final[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]


###################################################################################################################################

# Format the table 

pymoli_df_age_group_temp["Percentage of Players"] = pymoli_df_age_group_temp["Percentage of Players"].map("{:.2f}%".format)

pymoli_age_final["Average Purchase Price"] = pymoli_age_final["Average Purchase Price"].map("${:.2f}".format)

pymoli_age_final["Total Purchase Value"] = pymoli_age_final["Total Purchase Value"].map("${:.2f}".format)

pymoli_age_final["Avg Total Purchase per Person"] = pymoli_age_final["Avg Total Purchase per Person"].map("${:.2f}".format)

output5 = pymoli_df_age_group_temp

output6 = pymoli_age_final

# Output the results

output5
```
