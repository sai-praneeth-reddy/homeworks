```python
import json

import pandas as pd
```

```python
# Read the json file

f = open('samples.json',) 
```

```python
# import json file

data = json.load(f) 
```

```python
# Creating empty lists

otu_id_temp = []

sample_values_temp = []

otu_labels_temp = []
```

```python
for i in range(len(data['names'])):
    
    # Extract values from dictionary
    
    otu_id_extract = data['samples'][i]['otu_ids']
    
    sample_values_extract = data['samples'][i]['sample_values']
    
    otu_labels_extract =  data['samples'][i]["otu_labels"]
    
    # Appending to the empty list
    
    otu_id_temp.append(otu_id_extract)
    
    sample_values_temp.append(sample_values_extract)
    
    otu_labels_temp.append(otu_labels_extract)
```

```python
# Concatenate them into a single list

otu_id_list = [item for sublist in otu_id_temp for item in sublist]

sample_values_list = [item for sublist in sample_values_temp for item in sublist]

otu_labels_list = [item for sublist in otu_labels_temp for item in sublist]
```

```python
# Create a DF and extract top 10

Combine_df = pd.DataFrame(
    {'OTU_ID': otu_id_list,
     'Sample_Values': sample_values_list,
     'OTU_labels': otu_labels_list
    })
```
